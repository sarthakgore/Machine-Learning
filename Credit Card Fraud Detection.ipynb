# Importing Dependancies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Importing Dataset

```python
data = pd.read_csv(r"C:\Users\ADMIN\Downloads\archive\creditcard.csv")
```

# Distribuation of Transactions

# Seprating of data

```python
#Let us consider 0 --> Normal Transaction
    #            1 -->fraud Transaction
```

```python
legit = data[data.Class==0]
Fraud = data[data.Class==1]
```

```python
print(legit.shape)
print(Fraud.shape)
```

```python
#Stastical measures of data
```

```python
legit.Amount.describe()
```

```python
Fraud.Amount.describe()
```

```python
legit_sample=legit.sample(n=492)
```

```python
#Let's bulid a new dataset
```

```python
new_dataset= pd.concat([legit_sample,Fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset["Class"].value_counts()
```

# Splitting data into Features and Target

```python
x=new_dataset.drop(columns="Class",axis=1)
y=new_dataset["Class"]
```

```python
print(x)
```

```python
print(y)
```

# Split the data into Traning and Testing Data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# Now let's Train the Model

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#Training model
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
#Calculating Accuracy
```

```python
print("Accuracy on traing data:",training_data_accuracy)
```

```python
#Testing model
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
#Calculating Accuracy
```

```python
print("Accuracy score on test data:",test_data_accuracy)
```
